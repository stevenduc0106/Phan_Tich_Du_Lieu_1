```python
result=newdf.groupby(['KV'])[['T5','T6']].agg(['min','mean'])
print(result)
```

```python
#Pivod

pivot_result = pd.pivot_table( newdf,
                values=['T5','T6'],
                columns='KV',
                aggfunc=['min','mean','max']
              )
print(pivot_result)
```
