```python
#Explanatory Data Analysis (EDA)
# Mục đích: bao gồm các kỹ thuật khám phá dữ liệu
#(tức là từ dữ liệu rút trích các tri thức, hiểu biết)

#EDA => Insight
#NOTE: Để trích thông tin, tri thức từ dữ liệu, người ta thường dùng các đại lượng tổng hợp
#(Count, Min, Max, Sum, Mean, std...)
#hoặc biểu đồ trực quan(bar, pie,scatter,line, histogram...)

#Require : use dulieuxettuyendaihoc.csv

#Note: Cot, bien, thuoc tinh, dac trung = bien so

#phan 1 Thoa tac co ban

#Step 1: 
#2 Cat loc du lieu phu hop (Data Collection)
#3 Tong quan du lieu
#4 - Loai bo dong rong, dong trung , khao sat du lieu thieu -> xu ly du lieu thieu

import pandas as pd
import seaborn as sns
#Tai Du Lieu Len Data Frame, dat ten la df
df = pd.read_csv('dulieuxettuyendaihoc.csv',header = 0, delimiter = ',', encoding = 'utf-8')
```

```python
df = df[['T5','T6','GT','DT','KV','KT','NGONNGU','TOANLOGICPHANTICH','GIAIQUYETVANDE','NGAYTHI','DINHHUONGNGHENGHIEP']]
```

```python
df.info()
```

```python
df.rename(columns={'TOANLOGICPHANTICH': 'LOGIC', 
                   'GIAIQUYETVANDE': 'UNGXU', 
                   'DINHHUONGNGHENGHIEP': 'HUONGNGHIEP'}, inplace=True)
df.head(10)
```

```python
# Xoa Bo Cac Dong Du Lieu Rong
df.dropna(how='all')
#Xoa bo cac dong bi trung
df.drop_duplicates()
```

```python
#Dinh Tinh: Scalan, Mode, AI
#Dinh Luong: Scalan,mean(gia tri trung binh), media(gia tri trung vi), AI,
#Interpolate(Neu dung gia tri sau phan bi mat goi la BackWard, tri truoc phan bi mat la Forward )    
#Điền giá trị thiếu
df['DT'].fillna('KINH',inplace=True) #thực hiện thay thế tất cả các giá trị thiếu trong cột 'DT' bằng giá trị 'KINH'
df['DT']
#Scalan (Ve Tim Tren Mang Doc)
#Luu y: Với biến định tính ta có thể thay bằng giá trị yếu vị (mode)
#df['DT'].fillna(df['DT'].mode()[0],inplace=True)
```

```python

# Điền giá trị thiếu cho cột 'NGONNGU' bằng 0 và cập nhật DataFrame gốc
df['NGONNGU'].fillna(0, inplace=True)

# Điền giá trị thiếu cho cột 'LOGIC' bằng trung bình và cập nhật DataFrame gốc
df['LOGIC'].fillna(df['LOGIC'].mean(), inplace=True)

# Điền giá trị thiếu cho cột 'UNGXU' bằng trung vị và cập nhật DataFrame gốc
df['UNGXU'].fillna(df['UNGXU'].median(), inplace=True)

# Hiển thị một số dòng sau khi đã điền giá trị
print(df.head())

#NOTE: viới biến định lượng thì ta nên thay bằng trung vị
```

```python
# PHAN 2: Kỹ thuật Feature Engineering
#(thường dùng cho Machine Learning)
#Nếu chỉ là xử lý phân tíhc dữ liệu thì ta gọi là New Attril

#Đây là kỹ thuật tạo thêm hoặc biến đổi số liệu sẵn có thành các biến số mới 
#phù hợp nghiệp vụ phân tích

#Tạo biến TBTOAN: Trung Binh Toan Lop 12

#df['TBTOAN'] = (df['T5']+df['T6'])/2
df.loc[:, 'TBTOAN'] = (df['T5'] + df['T6']) / 2
df
```

```python
# Tạo biến XEPLOAI dựa trên TBTOAN
df.loc[df['TBTOAN'] < 5, 'XEPLOAI'] = 'FAIL'
df.loc[(df['TBTOAN'] >= 5) & (df['TBTOAN'] < 7), 'XEPLOAI'] = 'FAIR'
df.loc[(df['TBTOAN'] >= 7) & (df['TBTOAN'] < 9), 'XEPLOAI'] = 'GOOD'
df.loc[df['TBTOAN'] >= 9, 'XEPLOAI'] = 'EXCELLENT'

# Hiển thị một số dòng sau khi đã tạo biến XEPLOAI
print(df.head())
```

```python
#Tạo biến nhóm khối thi: NHOMKT thoã mãn
#A1: G1
#C : G3
#D1: G3
#A: G1
#B: G2
# Định nghĩa ánh xạ
dict_map = {'A1': 'G1', 'C': 'G3', 'D1': 'G3', 'A': 'G1', 'B': 'G2'}

# Tạo biến NHOMKT dựa trên KT
df.loc[:, 'NHOMKT'] = df['KT'].map(dict_map)

# Hiển thị cột NHOMKT
print(df['NHOMKT'])
```

```python
#Tao Bien So Diem COng: DIEMCONG

#Nếu NHOMKT thuộc nhóm G1, G2 và TBTOAN >= 5.0 thì 1.0
#ngược lại thì 0.0

def fplus(x, y):
    if (x == 'G1' or x == 'G2') and (y >= 5.0):
        return 1.0
    else:
        return 0.0

# Sử dụng lambda và map để áp dụng hàm fplus cho từng hàng dữ liệu và gán vào cột DIEMCONG
df['DIEMCONG'] = list(map(lambda x, y: fplus(x, y), df['NHOMKT'], df['TBTOAN']))

# Hiển thị cột DIEMCONG
print(df['DIEMCONG'])
df[['NHOMKT','TBTOAN','DIEMCONG']].head(5)
```
