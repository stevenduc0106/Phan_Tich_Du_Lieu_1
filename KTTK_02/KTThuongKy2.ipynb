```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('../../boston.csv',header=0,delimiter=',',encoding='utf-8')
```

```python
df
```

```python
#Tập dữ liệu đặc trưng
x = df[['AGE','LSTAT']].values #input
y = df[['MEDV']].values #output
```

```python
x
```

```python
#Chia tách ra trainning(tập huấn) và testting(kiểm thử)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train , y_test = train_test_split(x,y,test_size=0.25,random_state=16)
```

```python
print(X_train.shape) # kích thước của tập dữ liệu đặc trưng dùng huấn luyện
print(X_test.shape) # kích thước của tập dữ liệu đặc trưng dùng kiểm thử
print(y_train.shape) # kích thước của tập dữ liệu target dùng huấn luyện
print(y_test.shape) # kích thước của tập dữ liệu target dùng kiểm thử
```

```python
X_train
```

```python
y_train
```

```python
import numpy as np

# Định nghĩa hàm SiLU (Swish)
def silu(x):
    return x / (1 + np.exp(-x))

# Tạo một transformer áp dụng SiLU lên một mảng NumPy
class SiLUTransformer:
    def __init__(self):
        pass

    def transform(self, X):
        return silu(X)

# Sử dụng SiLU transformer
silu_transformer = SiLUTransformer()

# Áp dụng SiLU transformer lên dữ liệu
X_train = silu_transformer.transform(X_train)
```

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
#Scale cho inpput
sc = PolynomialFeatures(degree=2)
X_train = sc.fit_transform(X_train)
```

```python
X_train   # dữ liệu đặc trưng sau khi tiến hành transform
```

```python
y_train  # dữ liệu của tập dữ liệu target
```

```python
#Xây dựng mô hình hooid quy tuyến tính đa biến
# ? = A0 + A1*? + A2*? + epsilon
# điểm ngôn ngữ = A0 + T5-scale + T6-scale + epsilon
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X_train,y_train)
#
#
print(model.intercept_) # hệ số tự do 

print(model.coef_) # hệ số góc   bên trái là của age    bên phải là của lstat
#Sinh viên tự ghi lại mô hình
```

```python
X_train
```

```python
y_train
```

```python
#Đánh giá mô hình
import numpy as np
from sklearn import metrics
X_test = silu_transformer.transform(X_test)
X_test = sc.transform(X_test)

y_test_pred = model.predict(X_test)

print('Score or R-Squared:', model.score (X_test, y_test)) # giải thích phần trăm sự thay đổi của biến mục tiêu

print('Mean Absolute Error:',
        metrics.mean_absolute_error(y_test, y_test_pred)) # giá trị MAE

print('Mean Squared Error:',
        metrics.mean_squared_error(y_test, y_test_pred))  # giá trị MSE

print('Root Mean Squared Error:',
        np.sqrt(metrics.mean_squared_error(y_test, y_test_pred))) # giá trị RMSE
```
